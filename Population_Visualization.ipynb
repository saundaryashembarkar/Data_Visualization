```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['Indicator Name', 'Indicator Code'])
```

```python
population_2022 = df[['Country Name', '2022']]
population_2022 = population_2022.dropna()
population_2022.head()
```

```python
population_2022 = population_2022.sort_values(by='2022', ascending=False)
population_2022.head()
```

```python
top10 = population_2022.head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='Country Name', y='2022', data=top10)

plt.xticks(rotation=45)
plt.title("Top 10 Most Populated Countries (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(population_2022['2022'], bins=30, kde=True)

plt.title("Population Distribution (2022)")
plt.xlabel("Population")
plt.ylabel("Count")
plt.show()
```

Conclusion:

The bar chart shows that countries like India and China have the highest population in 2022.
The histogram indicates that most countries have relatively smaller populations,
while only a few countries have extremely high populations.
